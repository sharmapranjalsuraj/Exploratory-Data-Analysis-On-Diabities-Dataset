A heart attack, also known as a myocardial infarction (MI), is a serious medical emergency that happens when blood flow to the heart is blocked. This prevents the heart muscle from getting enough oxygen, which can cause it to die
 

#### Symptoms
- Chest pain, tightness, pressure, squeezing, or aching
- Pain that spreads to the arms, neck, jaw, or back
- Shortness of breath
- Feeling lightheaded or dizzy
- Sweating
- Nausea and vomiting
- Coughing or wheezing
- Anxiety that feels like a panic attack
